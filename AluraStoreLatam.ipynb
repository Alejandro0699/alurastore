### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
#Se sumo el total de cada tienda y se imiprimio el resultado.
total_tienda1 = tienda["Precio"].sum()
print(total_tienda1)
total_tienda2 = tienda2["Precio"].sum()
print(total_tienda2)
total_tienda3 = tienda3["Precio"].sum()
print(total_tienda3)
total_tienda4 = tienda4["Precio"].sum()
print(total_tienda4)


```

```python
#se genero un grafico de ventas totales por tienda.
Tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
Ventas = [total_tienda1, total_tienda2, total_tienda3, total_tienda4]

plt.bar(Tiendas, Ventas)
plt.title("Ventas Total por tienda")
plt.xlabel("Tiendas")
plt.ylabel("Ventas en $")
plt.show()
```

# 2. Ventas por categoría

```python
venta_categoria1= tienda.groupby("Categoría del Producto")["Precio"].sum().sort_values(ascending=False)
print(venta_categoria1)
venta_categoria2= tienda2.groupby("Categoría del Producto")["Precio"].sum().sort_values(ascending=False)
print(venta_categoria2)
venta_categoria3= tienda3.groupby("Categoría del Producto")["Precio"].sum().sort_values(ascending=False)
print(venta_categoria3)
venta_categoria4= tienda4.groupby("Categoría del Producto")["Precio"].sum().sort_values(ascending=False)
print(venta_categoria4)
```

# 3. Calificación promedio de la tienda


```python
calificacion_tienda1 = tienda["Calificación"].mean()
print(calificacion_tienda1)
califacion_tienda2 = tienda2["Calificación"].mean()
print(califacion_tienda2)
calificacion_tienda3 = tienda3["Calificación"].mean()
print(calificacion_tienda3)
calificacion_tienda4 = tienda4["Calificación"].mean()
print(calificacion_tienda4)
```

```python
#Calificacion promedio de los clientes
Tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
Calificacion = [calificacion_tienda1, califacion_tienda2, calificacion_tienda3, calificacion_tienda4]

plt.pie(Calificacion, labels=Tiendas, autopct='%1.1f%%', startangle=90)
plt.title("Calificación de las tiendas")
plt.show()
```

# 4. Productos más y menos vendidos

```python
#Tienda 1
producto_max1=tienda.groupby("Producto")["Precio"].count().idxmax()
cantidad_max=tienda.groupby("Producto")["Precio"].count()
cantidad1=cantidad_max[producto_max1]
print(f"La cantidad máxima de ventas de {producto_max1} es: {cantidad1}")
producto_min1=tienda.groupby("Producto")["Precio"].count().idxmin()
cantidad_min=tienda.groupby("Producto")["Precio"].count()
cantidad2=cantidad_min[producto_min1]
print(f"La cantidad mínima de ventas de {producto_min1} es: {cantidad2}")
# Tienda 2
producto_max2 = tienda2.groupby("Producto")["Precio"].count().idxmax()
cantidad_max2 = tienda2.groupby("Producto")["Precio"].count()
cantidad_t2_max = cantidad_max2[producto_max2]
print(f"La cantidad máxima de ventas de {producto_max2} es: {cantidad_t2_max}")

producto_min2 = tienda2.groupby("Producto")["Precio"].count().idxmin()
cantidad_min2 = tienda2.groupby("Producto")["Precio"].count()
cantidad_t2_min = cantidad_min2[producto_min2]
print(f"La cantidad mínima de ventas de {producto_min2} es: {cantidad_t2_min}")

# Tienda 3
producto_max3 = tienda3.groupby("Producto")["Precio"].count().idxmax()
cantidad_max3 = tienda3.groupby("Producto")["Precio"].count()
cantidad_t3_max = cantidad_max3[producto_max3]
print(f"La cantidad máxima de ventas de {producto_max3} es: {cantidad_t3_max}")

producto_min3 = tienda3.groupby("Producto")["Precio"].count().idxmin()
cantidad_min3 = tienda3.groupby("Producto")["Precio"].count()
cantidad_t3_min = cantidad_min3[producto_min3]
print(f"La cantidad mínima de ventas de {producto_min3} es: {cantidad_t3_min}")

# Tienda 4
producto_max4 = tienda4.groupby("Producto")["Precio"].count().idxmax()
cantidad_max4 = tienda4.groupby("Producto")["Precio"].count()
cantidad_t4_max = cantidad_max4[producto_max4]
print(f"La cantidad máxima de ventas de {producto_max4} es: {cantidad_t4_max}")

producto_min4 = tienda4.groupby("Producto")["Precio"].count().idxmin()
cantidad_min4 = tienda4.groupby("Producto")["Precio"].count()
cantidad_t4_min = cantidad_min4[producto_min4]
print(f"La cantidad mínima de ventas de {producto_min4} es: {cantidad_t4_min}")
```

```python
#Creando el gaficos para los productos mas y menos vendidos.
productos_max= (producto_max1, producto_max2, producto_max3, producto_max4)
cantidad_max= (cantidad1, cantidad_t2_max, cantidad_t3_max, cantidad_t4_max)
productos_min= (producto_min1, producto_min2, producto_min3, producto_min4)
cantidad_min= (cantidad2, cantidad_t2_min, cantidad_t3_min, cantidad_t4_min)

plt.bar(productos_max, cantidad_max, color='skyblue', label='Máximo')
plt.bar(productos_min, cantidad_min, color='orange', label='Mínimo')
plt.title("Productos más y menos vendidos")
plt.xlabel("Productos")
plt.ylabel("Cantidad de ventas")
plt.xticks(rotation=90)  # Opcional: rota los nombres para que no se superpongan
plt.tight_layout()       # Opcional: ajusta el gráfico para que todo entre bien
plt.legend()
plt.show()
```

# 5. Envío promedio por tienda

```python
precio_envio1= tienda["Costo de envío"].mean()
print(f"El precio promedio de envío de la tienda 1 es: {precio_envio1}")
precio_envio2= tienda2["Costo de envío"].mean()
print(f"El precio promedio de envío de la tienda 2 es: {precio_envio2}")
precio_envio3= tienda3["Costo de envío"].mean()
print(f"El precio promedio de envío de la tienda 3 es: {precio_envio3}")
precio_envio4= tienda4["Costo de envío"].mean()
print(f"El precio promedio de envío de la tienda 4 es: {precio_envio4}")
```

```python
#Creando el grafico para la media de envios
tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
envios = [precio_envio1, precio_envio2, precio_envio3, precio_envio4]

plt.pie(envios, labels=tiendas, autopct='%1.1f%%', startangle=90)
plt.title("Envios promedio por tienda")
plt.show()
```
